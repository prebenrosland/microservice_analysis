```python
import gzip
import pandas as pd
import numpy as np
from io import StringIO
from matplotlib import pyplot as plt

with gzip.open('mono_1_instance_1_constant_rate_results.gz', 'rb') as f:
    contentmono = f.read().decode('utf-8')
    
with gzip.open('micro_1_instance_1_constant_rate_results.gz', 'rb') as f:
    contentmicro = f.read().decode('utf-8')
        
#df_mono = pd.read_csv(StringIO(contentmono))
#df_micro = pd.read_csv(StringIO(contentmicro))

df_mono = pd.read_csv('mono_1_rate.csv')
df_micro = pd.read_csv('micro_1_rate.csv')
```

```python
values_mono = []
values_micro = []
average_mono = []
average_micro = []
timestamp_mono = []
timestamp_micro = []
for i in range(280):
    i+=10
    #print(df_micro["timestamp"].iloc[16*i + 1], df_micro["metric_name"].iloc[16*i + 1], df_micro["metric_value"].iloc[16*i + 1])
    values_mono.append(
        df_mono["metric_value"].iloc[16*i + 1]
    )
    #print(df["metric_name"].iloc[i])
    
for i in range(27):
    average_mono.append(np.mean(values_mono[i:i+9]))
    timestamp_mono.append((df_mono["timestamp"].iloc[16*i + 1] - df_mono["timestamp"].iloc[0])/100000)
```

```python
for i in range(280):
    i+=10
    values_micro.append(
        df_micro["metric_value"].iloc[16*i + 1]
    )
    
for i in range(27):
    average_micro.append(np.mean(values_micro[i:i+9]))
    timestamp_micro.append((df_micro["timestamp"].iloc[16*i + 1] - df_micro["timestamp"].iloc[0])/100000)
```

```python
plt.plot(timestamp_mono, average_mono, label='Monolith')
plt.plot(timestamp_micro, average_micro, label='Microservice')
#plt.plot(values_mono, label='Mono_1')
#plt.plot(values_micro, label='Micro_1')
plt.ylabel('y')
plt.xlabel('x')
plt.title('Request duration')
plt.ylim(50, 150)
plt.grid(True)
plt.legend()
plt.show()
```
